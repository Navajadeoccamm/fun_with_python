```python
# This file is part of the project's source code.
# Copyright (c) 2025 Daniel Monzon.
# Licensed under the MIT License. See the LICENSE file in the repository root.

# THIS JUPYTER NOTEBOOK PLOT A "HAPPY NEW YEAR" MESSAGE

import numpy as np
import matplotlib.pyplot as plt

# Data for the line segments
x_data = []
y_data = []

def add_line(x1, y1, x2, y2, steps=60):
    """Add a linear parametric segment to the path"""
    t = np.linspace(0, 1, steps)
    x = x1 + t * (x2 - x1)
    y = y1 + t * (y2 - y1)
    x_data.extend(x.tolist())
    y_data.extend(y.tolist())

def lift_pen():
    """Break the line for pen up (disconnected segments)"""
    x_data.append(np.nan)
    y_data.append(np.nan)

# Parameters
pos_x = 4  # starting margin
letter_width = 5
space_width = 3
height = 9
y_base = 2

# Simple block-style letter drawing functions
def draw_H():
    global pos_x
    add_line(pos_x, y_base, pos_x, y_base + height)
    add_line(pos_x, y_base + height // 2, pos_x + letter_width, y_base + height // 2)
    add_line(pos_x + letter_width, y_base, pos_x + letter_width, y_base + height)
    pos_x += letter_width + space_width

def draw_A():
    global pos_x
    add_line(pos_x, y_base, pos_x + letter_width // 2, y_base + height)
    lift_pen()
    add_line(pos_x + letter_width, y_base, pos_x + letter_width // 2, y_base + height)
    lift_pen()
    add_line(pos_x + letter_width // 4, y_base + height // 2, pos_x + 3 * letter_width // 4, y_base + height // 2)
    pos_x += letter_width + space_width

def draw_P():
    global pos_x
    add_line(pos_x, y_base, pos_x, y_base + height)
    add_line(pos_x, y_base + height, pos_x + letter_width, y_base + height)
    add_line(pos_x + letter_width, y_base + height, pos_x + letter_width, y_base + height // 2)
    add_line(pos_x, y_base + height // 2, pos_x + letter_width, y_base + height // 2)
    pos_x += letter_width + space_width

def draw_Y():
    global pos_x
    add_line(pos_x, y_base + height, pos_x + letter_width // 2, y_base + height // 2)
    lift_pen()
    add_line(pos_x + letter_width, y_base + height, pos_x + letter_width // 2, y_base + height // 2)
    lift_pen()
    add_line(pos_x + letter_width // 2, y_base + height // 2, pos_x + letter_width // 2, y_base)
    pos_x += letter_width + space_width

def draw_N():
    global pos_x
    add_line(pos_x, y_base, pos_x, y_base + height)
    add_line(pos_x, y_base + height, pos_x + letter_width, y_base)
    add_line(pos_x + letter_width, y_base, pos_x + letter_width, y_base + height)
    pos_x += letter_width + space_width

def draw_E():
    global pos_x
    add_line(pos_x, y_base, pos_x, y_base + height)
    add_line(pos_x, y_base + height, pos_x + letter_width, y_base + height)
    add_line(pos_x, y_base + height // 2, pos_x + letter_width, y_base + height // 2)
    add_line(pos_x, y_base, pos_x + letter_width, y_base)
    pos_x += letter_width + space_width

def draw_W():
    global pos_x
    add_line(pos_x, y_base + height, pos_x + letter_width // 4, y_base)
    add_line(pos_x + letter_width // 4, y_base, pos_x + letter_width // 2, y_base + height // 2)
    add_line(pos_x + letter_width // 2, y_base + height // 2, pos_x + 3 * letter_width // 4, y_base)
    add_line(pos_x + 3 * letter_width // 4, y_base, pos_x + letter_width, y_base + height)
    pos_x += letter_width + space_width

def draw_R():
    global pos_x
    add_line(pos_x, y_base, pos_x, y_base + height)
    add_line(pos_x, y_base + height, pos_x + letter_width, y_base + height)
    add_line(pos_x + letter_width, y_base + height, pos_x + letter_width, y_base + height // 2)
    add_line(pos_x, y_base + height // 2, pos_x + letter_width, y_base + height // 2)
    lift_pen()
    add_line(pos_x + letter_width // 2, y_base + height // 2, pos_x + letter_width, y_base)
    pos_x += letter_width + space_width

def draw_exclamation():
    global pos_x
    add_line(pos_x + letter_width // 2, y_base + height, pos_x + letter_width // 2, y_base + 3)
    lift_pen()
    add_line(pos_x + letter_width // 2, y_base + 1.5, pos_x + letter_width // 2, y_base + 0.5)
    pos_x += letter_width

# Build "Happy New Year!"
draw_H()
draw_A()
draw_P()
draw_P()
draw_Y()
pos_x += space_width * 2  # space between words
draw_N()
draw_E()
draw_W()
pos_x += space_width * 2
draw_Y()
draw_E()
draw_A()
draw_R()
draw_exclamation()

# Setup plot
fig, ax = plt.subplots(figsize=(16, 8), facecolor='black')
ax.plot(x_data, y_data, lw=4, color='gold')

# Festive touches
ax.set_title('Happy New Year 2026!', color='white', fontsize=24, pad=20)
ax.text(0.5, -0.1, 'Wishing you an amazing 2026!', transform=ax.transAxes,
        fontsize=16, color='white', ha='center')

# Auto-adjust and style
valid_x = [x for x in x_data if not np.isnan(x)]
valid_y = [y for y in y_data if not np.isnan(y)]
ax.set_xlim(min(valid_x) - 10, max(valid_x) + 10)
ax.set_ylim(min(valid_y) - 6, max(valid_y) + 8)
ax.set_aspect('equal')
ax.axis('off')
fig.patch.set_facecolor('black')

plt.tight_layout()
plt.show()
```
